```python
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt

from ipytools import slide, mplrc, hdisplay

import seaborn as sns

%matplotlib inline
%reload_ext export

pd.options.display.mpl_style = 'default'

data = pd.DataFrame(np.random.randint(0, 10, 100)).cumsum(0)
```

#An Example of Data

This is the cumulative growth over time

```python
data.plot();
```

#Plot with Seaborn

```python
with mplrc('figure.figsize', (12, 4)):
    fig, axes = plt.subplots(1, 2)
    sns.distplot(np.random.beta(20, 600, 1000), bins=50, ax=axes[0])
    sns.heatmap(np.random.randn(25).reshape((5, 5)), ax=axes[1])
    plt.tight_layout()
```

#Seaborn Continued

```python
with slide('You can do this with Seaborn, too!'):
    # Generate a random correlated bivariate dataset
    rs = np.random.RandomState(5)
    mean = [0, 0]
    cov = [(1, .5), (.5, 1)]
    x1, x2 = rs.multivariate_normal(mean, cov, 500).T
    x1 = pd.Series(x1, name="$X_1$")
    x2 = pd.Series(x2, name="$X_2$")

    # Show the joint distribution using kernel density estimation
    g = sns.jointplot(x1, x2, kind="kde", size=7, space=0)
```
